```python
#import library
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#data collection and analysis
#loading the data from csv file to a pandas dataframe
full_data = pd.read_csv('cust_data.csv')
full_data
x = full_data.head(50)
x
#print(full_data.columns)
# full_data.info()
```

# Groping

```python
a = input("Enter the brand name : ")

#optimum no.of cluster we choose
#training the k-means clustring model

df = x.drop(['Gender', 'Orders'], axis=1)
kmeans = KMeans(n_clusters=3)
kmeans.fit(df)

#visualizing the cluster
plt.scatter(df['Cust_ID'], df[a], c=kmeans.labels_)
plt.show()
```

# plotting the elbow graph

```python
#finding g,h value for different number of cluster
g = []
h = []

for i in range(1, 11):
    km = KMeans(n_clusters=i)
    km.fit_predict(df)
    g.append(km.inertia_)
    h.append(i)
plt.plot(g,h)
plt.show()
```
